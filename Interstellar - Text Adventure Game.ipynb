```python
"""
Created on Tuesday Jan 25 22:06:56 2022

@author: siddharth.varada
"""

"""
    DocString:
    
    A) Introduction:
    This game is based on the movie Interstellar (2014) by Christopher Nolan.
    It consists of four stages, and also has defined functions for start, 
    success, and failure. This game is based on the plot of the movie. A 
    reading on https://en.wikipedia.org/wiki/Interstellar_(film) will help 
    although you are better off avoiding it to keep the game fresh and fun.
    
    Round 1: Start by answering a basic question.
    Round 2: Choose your plan to save humanity.
    Round 3: Choose your path to move ahead.
    Round 4: Make your choice.
    
    B) Known Bugs and/or Errors:
    While the game accounts for variations in entries, it assumes that the 
    user doesn't make any spelling errors and does not use incorrect spacing.
    Doing so might lead to failure even if the answer was known to the user.
     
"""
#importing system, random and time packages
import sys
import random
import time

#The game starts from this function
def start():
    print("""
         Welcome to the Interstellar Text Adventure Game. Are you ready to 
         enter the world of Christopher Nolan?
         """)
    input(prompt = "<Press enter to continue>\n")
    welcome() #moves to welcome stage

def welcome(): #This is the first stage
    #The lists allow for variations in text entries based on any particular
    #option selected
    lst1 = ["global", "warming", "global warming"]
    lst2 = ["crop", "blight", "blights", "crop blight", "crop blights", "dust", "storm", "storms", "dust storm", "dust storms", "crop blights and dust storms"]
    lst3 = ["farm", "farming"]
    lst4 = ["no", "nothing", "no problems"]
    print("What are the problems suffered by humanity shown in the movie?\n")

    print("""
    Press 1 for global warming.
    Press 2 for crop blights and dust storms.")
    Press 3 for farming.")
    Press 4 for no problems
    """)
    choice = input(prompt = "> ")
    choice = choice.casefold() #accounting for case sensitivity
    
    #conditional statement for option 1
    if choice == '1' or choice in lst1:
        time.sleep(0.5)
        print("""
        It's not the right answer but appreciate your awareness of a 
        real world problem, please answer questions.
        """)
        #It is the wrong answer but a new set of questions created to reward 
        #the user
        reward_list = [ "What is Cooper's son's name?", "What is the name of the spacecraft used", "What is the concept of science used to explain variations in time in different planets?", "Name the robot used on the spacecraft", "Who is Amelia Brand's love interest", "Cooper reunites with Murphy at a space habitat. Around which planet does it orbit?"]
        #the below list stores the answers
        reward_ans = ["tom", "endurance", "relativity", "tars", "edmund", "saturn"]
        questions = 3 #Using a list of 3 questions from the 6 available
        random_num = random.randint(1,6) #generates a random integer from 1 to 6
        selection = input(prompt = "Guess a number from 1 to 6\n") #inputs user guess
        
        while questions > 0: #will execute until the question count reaches 0
            #random generation of questions
            question = random.choice(reward_list) #picks a random question from the list
            input(prompt = "<Press enter for your question>\n")
            
            try:# executes until error occurs
                selection = int(selection)
                if selection == random_num: #user gets one question if guess is right
                    print("Answer one question.\n")
                    question = reward_list[random_num-1] #generating the question
                    print(question) #printing question
                    ans = input(prompt = "Enter here:\n")

                    if ans == reward_ans[random_num-1]:
                        print("You got it right!")
                        planning()

                    else:
                        fail()

                else: #user gets three question if guess is wrong
                    print(question)
                    if question == reward_list[0]: #if user gets the 1st question
                        name = input(prompt = "Enter here: \n")

                        if name.casefold() == 'tom': 
                            #casefold using within each nested conditional to account
                            #for case sensitivity
                            print("You got it right!")
                            questions -= 1 #decrementing the question count

                        else:
                            fail()

                    elif question == reward_list[1]: #if user gets the 2nd question
                        spacecraft = input(prompt = "Enter here: \n")

                        if spacecraft.casefold() == 'endurance':
                            print("You got it right!\n")
                            questions -= 1

                        else:
                            fail() #answering it wrong leads to failure

                    elif question == reward_list[2]: #if user gets the 3rd question
                        concept = input(prompt = "Enter here: \n")

                        if concept.casefold() == 'relativity':
                            print("You got it right!")
                            questions -= 1

                        else:
                            fail()

                    elif question == reward_list[3]: #if user gets the 4th question
                        robot = input(prompt = "Enter here: \n")

                        if robot.casefold() == 'tars':
                            print("You got it right!\n")
                            questions -= 1

                        else:
                            fail()

                    elif question == reward_list[4]: #if user gets the 5th question
                        love = input(prompt = "Enter here: \n")

                        if love.casefold() == 'edmund':
                            print("You got it right!\n")
                            questions -= 1

                        else:
                            fail()

                    elif question == reward_list[5]: #if user gets the 6th question
                        planet = input(prompt = "Enter here: \n")

                        if planet.casefold() == 'saturn':
                            print("You got it right!\n")
                            questions -= 1

                        else:
                            fail()

                    else: #if user selects from oustide the given options
                        print("Invalid entry. Please try again.\n")
                        input(prompt = '<Press any key to continue>\n')
                        welcome()
            
            except: #catching non-numeric entries
                print("Non-numeric entry not allowed")
                welcome()  
        
        print("You made it to the next stage!\n")
        input(prompt = "<Press enter to continue>\n")
        planning()
        
    elif choice == '2' or choice in lst2: #conditional statement for option 2
        time.sleep(0.5) #kernel suspends execution for half a second
        print("""
        It is the year 2067. Humanity is fighting for survival due to the 
        presence of crop blights and dust storms. Farming is not sustainable 
        anymore. Cooper and his daughter Murphy are looking for something 
        better to do.
        """)
        time.sleep(0.1)
        #lists accounting for variation in entries for next question
        lsta = ["geographic", "geographic coordinates", "binary", "binary code", "geographic coordinates in binary", "geographic coordinates in binary code"]
        lstb = ["ghosts", "cupboard", "ghosts in the cupboard", "ghosts in the cupboard"]
        lstc = ["aliens", "communicate", "aliens trying to communicate"]
        print( 
        """
        What did Cooper manage to decipher from the dust storms in his daughter Murphy's room?
          1. geographic coordinates in binary code   
          2. ghosts in the cupboard
          3. aliens trying to communicate
        """)
        
        choice = input(prompt = "> ")
        choice = choice.casefold() #accounting for case sensitivity
        
        if choice == '1' or choice in lsta: #if user chooses option 1
            print("That's correct. Let's proceed further\n")
            planning()

        elif choice == '2' or choice in lstb: #if user chooses option 2
            print("That's incorrect!")
            fail()

        elif choice == '3' or choice in lstc: #if user chooses option 3
            print("That's incorrect!")
            fail()

        else:
            print("Invalid entry. Please try again.\n")
            input(prompt = '<Press any key to continue>\n')
            welcome()            
    
    elif choice == '3' or choice in lst3: #if user selects option 3
        print("That's the wrong answer. You haven't watched the movie, have you?")
        fail()
        
    elif choice == '4' or choice in lst4: #if user selects option 4
        print("Come on, you can do better!")
        fail()    
    
    else: #if user selects from outside the given options
        print("Invalid entry. Please try again.\n")
        input(prompt = '<Press any key to continue>\n')
        welcome()
    


def planning(): #This is the planning stage
    #lists allowing for variation in text entries in the planning stage
    lst5 = ["anti-gravity", "anti-gravitational", "propulsion", "propulsion theory", "anti-gravitational propulsion theory" "propel settlements", "propel settlements into space", "developing an anti-gravitational propulsion theory to propel settlements into space"]
    lst6 = ["launching", "launch", "endurance", "endurance spacecraft", "carrying 5,000 frozen human embryos", "carrying 5,000 embryos", "carrying frozen embryos", "5000", "embryos", "launching the Endurance spacecraft carrying 5,000 frozen human embryos to settle a habitable planet"]
    lst7 = ["scared", "too scared"]
    time.sleep(0.5)
    print("""
    Cooper follows the geographical candidates to a secret NASA facility led
    by Professor Brand. A few decades ago, a wormhole near Saturn opened up a
    path to a distant galaxy with twelve planets having potential living conditions for human
    beings. Three volunteers who surveyed went to those planets 
    reported positive results. Brand gives two plans for humanity's survival.
    """)
    input(prompt = '<Press any key to continue>\n')
    
    print("""
    Which plan do you choose for saving humanity?
    Press 1 for developing an anti-gravitational propulsion theory to propel settlements into space
    Press 2 for launching the Endurance spacecraft carrying 5,000 frozen human embryos to settle a habitable planet
    Press 3 if you are too scared
    """)
    
    choice = input(prompt = "> ")
    choice = choice.casefold() #accounting for case sensitivity
    
    #conditionals based on what option the user selects
    if choice == '1' or choice in lst5:
        print("""
        You have taken the safe choice! You don't seem the adventurous 
        type.\n""")
        print("""Now that you have chosen this, answer three questions. You
        still can't avoid the adventure later though!
              
        """)
        #the list results stores the answers for the questions
        results = ["infinite", "inertial frame of reference", "(mo)^2(c)^4"] 
        time.sleep(0.1)
        print("""
        1. As an object approaches the speed of light, it’s mass becomes
        __________. (a) Zero, (b) Double, (c) Remains Same, (d) Infinite
        """)
        ans1 = input(prompt = "Enter here\n") #taking user input 1
        if ans1.casefold() == results[0] or ans1 == 'd' or ans1 == '(d)' or ans1 == '(d) infinite':
            pass #moves to 2nd question
        else:
            fail()
            
        print("""
        2. According to Einstein’s Special Theory of Relativity, laws of 
        physics can be formulated based on ____________. (a) inertial frame of 
        reference, (b) non-inertial frame of reference, (c) both inertial and 
        non-inertial frame of reference, (d) quantum State
        """)
        ans2 = input(prompt = "Enter here\n") #taking user input 2
        if ans2.casefold() == results[1] or ans2 == 'a' or ans2 == '(a)' or ans2 == '(a) inertial frame of reference':
            pass #moves to next question
        else:
            fail() #leads to failure
        print("""     
        3. For Einstein’s relation, E2 – p2c2 = _____________. (a) mo(c)^2, 
        (b) (mo)^2(c)4, c) mo(c)^4, d) (mo)^2(c)^6
        """)
        ans3 = input(prompt = "Enter here\n") #taking user input 3
        if ans3.casefold() == results[2] or ans3 == 'b' or ans3 == '(b)' or ans3 == '(b) (mo)^2(c)4':
            path() #moves to next stage
        else:
            fail()

    elif choice == '2' or choice in lst6:
        path()
        
    elif choice == '3' or choice in lst7:
        print("You shoudn't have chosen this adventure if you are so scared!")
        fail()
    
    else: # if user selects outside the given options
        print("Invalid entry. Please try again.\n")
        planning()



def path(): #This is the 3rd stage
    #lists allowing for variation in text entries in the path stage
    lst8 = ["edmund", "edmund's", "edmund's planet", "edmund planet"]
    lst9 = ["mann", "mann's", "mann's planet", "mann planet"]
    lst10 = ["go", "go back", "go home", "back home", "go back home", "want to go back home", "want to go home", "want to go", "you want to go back home", "want", "you want", "i want"]
    time.sleep(0.5)
    print("""
    Are you ready for the exciting journey ahead? Cooper, Romilly, Doyle and
    Amelia leave to survey the three planets! Unfortunately, they don't 
    account for the proximity of Miller's ocean-world planet to the black hole
    Gargantua which costs them a time equivalent to 23 years on Earth. Now 
    they have to make a choice between Mann's planet and Edmund's planet.
    """)
    time.sleep(0.1)
    print("""
    Edmund's planet stopped sending transmissions but has better potential 
    according to Amelia. Mann's planet is still sending transmissions. 
    Amelia's choice may be biased by her love for Edmund.
    """)
    #New question to answer
    print("Which planet do you want to go to?\n")
    print("""
    Press 1 for Edmund's planet
    Press 2 for Mann's planet
    Press 3 if you want to go back home
    """)
    
    choice = input(prompt = "> ")
    choice = choice.casefold() #accounting for case sensitivity
    
    if choice == '1' or choice in lst8: #if user selects option 1
        print("Well done! You made the right decision. Love has won!")
        dilemma()

    elif choice == '2' or choice in lst9: #if user selects option 2
        print("Sorry, Mann has betrayed you! Game over.")
        fail()

    elif choice == '3' or choice in lst10: #if user selects option 3
        print("It's a pity you want to go home now after coming this far!")
        fail()

    else: #if user's input doesn't match with the options
        print("Invalid entry. Please try again.\n")
        path()

def dilemma(): #This is the last stage
    #lists allowing for variation in text entries in the dilemma stage
    lst11 = ["staying", "back", "staying back"]
    lst12 = ["going", "going to", "to amelia", "going amelia", "going to amelia"]
    lst13 = ["if", "undecided", "if undecided"]
    time.sleep(0.5)
    print("""
    Cooper is reunited with his daughter in a space habitat orbiting Saturn.
    Sad news though, she is now 89 years old and in her last breaths.
    """)
    time.sleep(0.1) #small suspension of kernel execution to give breathing space
    print("Should Cooper stay with Murphy or go to Amelia?\n")
    print("""
    Press 1 for staying back
    Press 2 for going to Amelia
    Press 3 if undecided
    """)
    
    choice = input(prompt = "> ")
    choice = choice.casefold() #accounting for case sensitivity
    
    if choice == '1' or choice in lst11:
        print("""Sorry! While wanting to stay with Murphy is 
        understandable, it costs him valuable time!""")
        fail()

    elif choice == '2' or choice in lst12:
        print("Great job! This planet has lot of potential for human life!")
        humanity()

    elif choice == '3' or choice in lst13:
        print("It's a pity you want to go home now after coming this far!")
        fail()

    else:
        print("Invalid entry. Please try again.\n")
        dilemma()

def fail(): #This function is entered if the user fails at any point
    print('You have failed in your mission to save humanity.\n')
    input(prompt = 'Game Over! Press any key to exit.\n')
    quit(0) #stops execution

def humanity(): #This is used when the user successfully completes the game.
    print("""
    You have succeeded in your mission to save humanity! Great job!
    """)
    input(prompt = "<Well done, champion! Press any key to exit>\n")
    quit(0) #stops execution

start() #Tells the kernel where to start execution
```
